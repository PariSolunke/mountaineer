```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn import ensemble
from gale import create_mapper

data = load_breast_cancer()

X= pd.DataFrame(data.data, columns = data.feature_names)

y = data.target
column_names=data.feature_names
model = ensemble.IsolationForest(random_state=2022)
model.fit(X)
preds=model.predict(X)
lens1 = model.decision_function(X)
lens2 = X['mean radius'].to_numpy()
X = X.to_numpy()
```

```python
#create list of mapper outputs and lenses
mapper_outputs=[]
mapper1 = create_mapper(X, lens1, resolution=10, gain=0.3, dist_thresh=0.5)
mapper2 = create_mapper(X, lens2, resolution=10, gain=0.3, dist_thresh=0.5)

mapper_outputs=[mapper1, mapper2]
lenses = [lens1, lens2]
```
